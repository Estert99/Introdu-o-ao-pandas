No codigo abaixo eu:
Importei a os dados fiz o pre processamento.
Identifiquei o tipo de dados.
Renomei as colunas
Removi os valores nulos.



```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("/Projeto.xlsx")

print(df.head()) #imprime as 5  primeiras linhas da tabela

print(df.info()) #iprime o nome da coluna e os tipos de dados

print(df.describe()) #imprime o resumo estatistico

print(df.isnull().sum()) #checa  os valores nulos

print(df["Categoria"].unique()) #imprime as categorias unicas

print(df["Tipo"].value_counts()) #imprime quantos regristo de cada tipo

df= df.dropna(subset = ['Data','Valor ']) #elima os valores nulos
df.columns = df.columns.str.replace(' ', '_') #renomei a coluna Valor
df['Tipo']= df.Tipo.replace('Saida',"Saída") #renomei os dados da coluna Tipo

```
